## Handling DateFrame

```python
players_with_val=pd.merge(players, players_valuations, on='player_id',how='inner') # 두 데이터프레임 병합, on='player_id' 기준으로 내부 조인, how는 
players_with_val[players_with_val['last_name']=='Son'].tail()
```

```python
add=lambda x,y: x+y
num_a=5
num_b=10

print(add(num_a, num_b))
```

```python
players_with_val['dateyear']= players_with_val['date'].apply(lambda x: int(x[:4])) # 년도 추출하여 새로운 열로 추가, 람다함수는 간단한 함수를 한 줄로 표현할 때 사용, 여기서는 문자열에서 년도 부분을 추출하여 정수로 변환하는 함수 정의
```

```python
players_with_val['age']= players_with_val['dateyear'] - players_with_val['date_of_birth'].apply(lambda x: int(x[:4]))
```

```python
#중복값제거 

players_with_val.drop_duplicates(['player_id','dateyear'], keep='last', inplace=True)  
players_with_val[players_with_val['last_name']=='Son']
#[]기준으로 가져오겠다. last는 중복값중에서 마지막값만 가져오겠다
```
