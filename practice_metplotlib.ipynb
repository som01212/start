## 멧플롯립

## import Matplotlib

```python
import matplotlib.pyplot as plt
```

## make Data

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [30, 50, 25, 40]

x_values =[1, 2, 3, 4,5]
y_values = [10,20,15,25,30]
```

## try various function

```python
plt.bar(categories, values) #막대 그래프 생성
plt.xlabel('Categories') #x축 라벨
plt.ylabel('Values') #y축 라벨
plt.title('Bar Chart Example') #그래프 제목
plt.show() #그래프 표시
```

```python
plt.plot(x_values, y_values, marker='x', linestyle='--') #선 그래프 생성 marker='x'생성 점 스타일 지정 linestyle='--'선 스타일 지정
plt.xlabel('X Values') #x축 라벨
plt.ylabel('Y Values') #y축 라벨
plt.title('Line Chart Example') #그래프 제목
plt.grid(True) #격자 표시
plt.show() #그래프 표시
```

```python
import numpy as np

x_scatter=np.random.rand(50) #x축 데이터 생성  난수들이 무작위로 50개 생성
y_scatter=np.random.rand(50) #y축 데이터 생성 난수들이 무작위로 50개 생성
colors_scatter=np.random.rand(50) #색상 데이터 생성 난수들이 무작위로 50개 생성
sizes_scatter= np.random.randint(10,100,50) #크기 데이터 생성 10에서 100사이의 정수 난수들이 무작위로 50개 생성

labels_pie=['category A','category B','category C','category D'] #파이 차트 라벨
sizes_pie=[25,35,20,20] #파이 차트 크기
```

```python
np.random.rand(50) # 무작위로 50개의 난수 생성
```

```python
plt.scatter(x_scatter, y_scatter, c=colors_scatter, s=sizes_scatter, alpha=0.6) #산점도 생성, c=colors_scatter 색상 지정, s=sizes_scatter 크기 지정, alpha=0.6 투명도 설정
plt.xlabel('X Values') #x축 라벨
plt.ylabel('Y Values') #y축 라벨
plt.title('Scatter Plot Example') #그래프 제목
plt.colorbar(label='Color Scale') #색상 막대 표시
plt.show() #그래프 표시
```

```python
plt.pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90) #파이 차트 생성  labels=labels_pie 라벨 지정, autopct='%1.1f%%' 퍼센트 표시, startangle=90 회전 각도 설정
plt.axis('equal') #파이 차트를 원형으로 표시
plt.title('Pie Chart Example') #그래프 제목
plt.show() #그래프 표시
```
